```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
a=sns.load_dataset("movies")
sns.catplot(x="year",y="votes",kind="violin",data=a)
```

```
# Handiling missing values (Dealing with NA)
#Find for NA's
movies.isna() # to find for NA's in a dataframe
```

```
#Fill NA's with zero in dataset
movies.fillna(0) 
```

```
#Fill NA's with previous value in data set
movies.fillna(method='pad')
```

```
#Fill NA's with next values in dataset
movies.fillna(method ='bfill')
```

```
#Fill NA's with a constant/text 
movies.fillna("Not available",inplace=True) # Note: this will modify any other views on this object
movies
```

```
#Use interpolate function to fill the missing values which are in data set
movies.interpolate(method="linear") #‘linear’: ignore the index and treat the values as equally spaced. default
```

```
#Remove rows with NA's in dataset
movies.dropna() # Drop rows with NA
```

```
#Remove row which contains all NA's
movies.dropna(how="all") # Drop the rows containing all NA's
```

```
#Remove the columns containing NA's
movies.dropna(axis=1) # drop columns with NA's
```
